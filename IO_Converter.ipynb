## Vocoder

```python
import numpy as np
data = np.fromfile('qnn_calibration/vocoder/latent_100.raw', dtype=np.float32)
len(data) // 256
# 144

# Load ONNX-style input (NCHW: 1, 80, 2048)
nchw = data.reshape(1, 144, 256)

# Transpose to NHWC (1, 2048, 80) for DLC
nhwc = nchw.transpose(0, 2, 1)

print(f"ONNX shape (NCHW): {nchw.shape}")   # (1, 144, 256)
print(f"DLC shape (NHWC): {nhwc.shape}")    # (1, 256, 144)

# Save for SNPE
nhwc.astype(np.float32).tofile('./Input/vocoder_gpu.raw')
```

```python
import numpy as np
from scipy.io import wavfile

# Read audio
sr, audio = wavfile.read("./output/output.wav")

# Convert to float for safety (optional but recommended)
audio = audio.astype(np.float32)

# Time boundaries (seconds)
start_sec = 1.0
end_sec = 6.0

start_sample = int(start_sec * sr)
end_sample = int(end_sec * sr)

# Trim
trimmed_audio = audio[start_sample:end_sample]

# Write output
wavfile.write("./output/oP_trim.wav", sr, trimmed_audio.astype(np.float32))
```
